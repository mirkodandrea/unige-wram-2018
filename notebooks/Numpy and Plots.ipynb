## Numpy
Dealing with vectors and matrices efficiently requires the **numpy** library. 

In order to use a library in our program, we first have to tell python that we need it. This is done with the import statement:
```python
import numpy
```
#### A note on importing libraries 
A library (also called **module**) needs to be installed in your system before you can use it. This environment comes prepackaged with the needed libraries, so you don't need to worry about the installation process. If you want to learn how to install a module, head to the [official guide](https://docs.python.org/3/installing/index.html)

While importing a module, can also rename it, e.g. by using a shorter name:
```python
import numpy as np
```

Or import only a subset of the provided functions from the module:
```python
from numpy import array
```

````python
# Let's import the library
import numpy as np
````

Now we can use the library. Let's first declare a couple of vectors:

````python
x = np.array([1,2,3,4,5])
y = np.array([2*i for i in x])
````

And apply some element wise operations

````python
x+y, x-y, x*y, x/y, x**2 # element wise operations
````

We can subset the data, using the slicing operator or a conditional subsetting.

````python
print("first three elements of x", x[0:3])
print("elements of x >3", x[x>3])
````

Let's apply a reshape function to a subset of array x.

````python
X = x[:4].reshape(2,2) # reshape to 2x2 matrix
2*X # multiply by a scalar
````

Numpy includes a lot of useful functions, you can explore the documentation at https://docs.scipy.org/doc/numpy-1.13.0/reference/

Numpy supports arrays and matrices with many of the features that would be familiar to matlab users. See here quick summary of [numpy for matlab users](https://docs.scipy.org/doc/numpy-dev/user/numpy-for-matlab-users.html).

# Plotting
There are several libraries that can be used for plotting data.
The **matplotlib** library is the most used. It was developed as a port of the Matlab plot subsystem for python, so it share a similar interface and function names. 
For more information see [http://matplotlib.org/users/beginner.html]

To get started and use the library in a jupyter notebook, we need to import the library and to call the **ion** function that will set the library to show inline output for our plot functions. This is necessary because we are using a Jupyter notebook.

````python
import matplotlib.pyplot as plt
plt.ion() #set inline plot
plt.style.use('ggplot') # set the style of the plots
````

Now we can try something simple:

````python
x = np.linspace(0, 10, 100)
y = x**2
plt.plot(x, y)
plt.ylabel('x^2')
plt.xlabel('x value')
plt.show() # this is implicit in the notebook environment
````

It is possible to set the axis limit, labels, title, using the xlim, ylim, xlabel, ylabel functions, similarly to matlab.

We can define a new figure, and specify some properties, using the **figure** function.
All the plot function call after a figure definition will be plotted in the same graph. We can specify the plot colour and add a legend.

````python
# A slightly more complicated plot with the help of numpy
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

# let's create a figure
fig = plt.figure(figsize=(10, 5)) 

# we add a label to show it in the legend
plt.plot(x, np.sin(x), color='b', label='sin(x)')

# set line properties
plt.plot(x, np.cos(x), 
             color='r', 
             label='cos(x)',
             linewidth=2.5, 
             linestyle="-"
        )

plt.legend()
````

Similarly to Matlab, you can draw multiple plots on the same figure with the `plt.subplot` function

````python
x = np.linspace(-1, 1, 100)
y1 = np.arctan(x)
y2 = np.tan(x)

plt.figure(figsize=(6, 10))
plt.subplot(2, 1, 1)
plt.title('tan(x)')
plt.plot(x, y1)
plt.subplot(2, 1, 2)
plt.title('arctan(x)')
plt.plot(x, y2)
plt.show()
````

It is also possible to plot two or more series togheter, defining a 2d array of y values

````python
x = np.linspace(-3, 3, 100)
data = np.array([
    x**2, 
    x**3
])
plt.plot(x, data.T)
plt.legend(['x^2', 'x^3'])
plt.show()
````

It is also possible to plot a 2d matrix, using the `imshow` or `pcolor` functions, and plot isolines using `contour`.
The parameter **cmap** is passed to use a colormap, which will determine how to assign a color to each value of the input matrix.


````python
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(X)**2 + np.sin(Y)**2

plt.pcolor(X, Y, Z, cmap='viridis')
plt.colorbar()

plt.contour(X, Y, Z, cmap='gray')

plt.show()
````
